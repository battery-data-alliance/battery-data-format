# Ingest raw data into BDF

This notebook demonstrates `bdf.ingest`, which:

- converts raw vendor files into BDF artifacts
- validates existing BDF files
- writes outputs to parquet by default (csv optional)


```python
from pathlib import Path
import bdf
```

## Pick a source directory

By default this uses the local `data/` folder in the repo.

```python
source_dir = Path("../data")
source_dir.exists()
```

## Run ingest

Outputs are written to `out/ingest/` so the raw data stays untouched.

```python
out_dir = Path("out/ingest")
summary = bdf.ingest(
    source_dir,
    out_dir=out_dir,
    format="parquet",
    recursive=True,
    validate_existing=True,
    validate_converted=True,
    raise_on_error=False,
)
summary
```

## Inspect results

```python
print("converted:", len(summary.get("converted", [])))
print("validated:", len(summary.get("validated", [])))
print("failed:", len(summary.get("failed", [])))
print("skipped:", len(summary.get("skipped", [])))
```

```python
summary.get("failed", [])[:5]
```

## Validate-only pass

If you just want to validate existing BDF artifacts, point `ingest` at a folder
that already contains BDF files and disable conversion validation.

```python
summary_validate_only = bdf.ingest(
    out_dir,
    format="parquet",
    recursive=True,
    validate_existing=True,
    validate_converted=False,
)
summary_validate_only
```

## Notes

- For `.nda/.ndax` files, install `bdf[nda]` (NewareNDA) or `bdf[plotly]` if you
  want Plotly for interactive visualization.
- Use `format="csv"` if you need `.bdf.csv` output instead of parquet.
