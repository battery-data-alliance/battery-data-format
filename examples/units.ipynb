```python
from bdf.units import resolve_unit, convert, ureg
import pandas as pd

print(resolve_unit("Voltage / V", as_string=True))            # V
print(resolve_unit("AhAccu#Ah", as_string=True))              # A*h
print(resolve_unit("specific_capacity_milliampere_hour_per_gram", as_string=True))   # from spec MR name: W
print(resolve_unit("https://w3id.org/battery-data-alliance/ontology/battery-data-format#discharging_capacity_ah", as_string=True))   # from spec MR name: W

s = pd.Series([3.6, 3.7, 3.8], name="Voltage / V")
s_mV = convert(s, "mV", from_unit=None)  # infer 'V' from the label
print(s_mV.head())  # 3600., 3700., 3800.æ

ureg.Unit("milliampere*hour/gram")

```
